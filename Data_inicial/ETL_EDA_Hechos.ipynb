## Modificacion de Type


#### Los valores de la columna ID los convierto a int

```python
df_siniestros['ID'] = df_siniestros['ID'].str.replace('-','')
df_siniestros['ID'] = df_siniestros['ID'].fillna(0)
df_siniestros['ID'] = pd.to_numeric(df_siniestros['ID'], errors='coerce').astype(int)
```

```python
df_victimas['ID_hecho'] = df_victimas['ID_hecho'].str.replace('-','')
df_victimas['ID_hecho'] = df_victimas['ID_hecho'].fillna(0)
df_victimas['ID_hecho'] = pd.to_numeric(df_victimas['ID_hecho'], errors='coerce').astype(int)
```

```python
df_victimas.columns
```

```python
df_siniestros.columns
```

```python
df_victimas.drop(columns={'AAAA','MM','DD','FECHA','VICTIMA'},inplace=True)
```

___
### Hago la union de los dos dataframes

```python

df_hechos = pd.merge(df_siniestros, df_victimas, left_on='ID', right_on='ID_hecho', how='inner')
```

```python
df_hechos
```

## Graficos de exploracion de la informacion

### Cuando razono acerca de siniestros viales, me hago una idea de quienes pueden ser los que participan en ellos, pero para tener mas claro la informacion de mi dataframe, voy a graficar unos puntos que me resultan interesantes como:
- Quienes son aquellos que más participan (Tanto victimas, como acusados, como el par en el siniestro vial, como en total)
- En que tipos de calles ocurren mas accidentes
- Que año tuvo mas siniestros viales
- Si hay momentos en el año que ocurran más accidentes
- En qué comuna han ocurrido mas siniestros
- Que participantes predominan cada comuna
- La hora en la que mas se frecuentan los siniestros
- Si hay alguna relacion entre hora y participantes

Entre otras cosas
___

### **PARTICIPACION** en siniestros

```python
df_hechos['ACUSADO'].unique()
```

```python
df_hechos['VICTIMA'].unique()
```

```python
# Participantes (Víctimas, Acusados, Total)
participantes = ['VICTIMA', 'ACUSADO']

plt.figure(figsize=(15, 8))

for participante in participantes:
    counts = df_hechos[participante].value_counts()
    plt.bar(counts.index, counts.values, alpha=0.7, label=participante)

plt.title('Cantidad de Veces que Aparecen los Participantes en Siniestros Viales')
plt.xlabel('Participantes')
plt.ylabel('Cantidad')
plt.legend()
plt.show()
```

___
## **Calles** con mas siniestros

```python
plt.figure(figsize=(12, 3))
sns.countplot(y='TIPO_DE_CALLE', data=df_hechos)
plt.title('Tipos de Calle en Accidentes')
plt.xticks(rotation=45)
plt.show()
```

___
## Siniestros viales por **AÑO**

```python
plt.figure(figsize=(10, 5))
sns.lineplot(x='AAAA', y='ID', data=df_hechos.groupby('AAAA').count().reset_index())
plt.title('Número de Siniestros Viales por Año')
plt.show()
```

___
## **MES** con mas siniestros 


```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='MM', y='ID', data=df_hechos.groupby('MM').count().reset_index())
plt.ylabel('Cantidad de accidentes')
plt.xlabel('Mes')
plt.title('Meses con Mayor Número de Accidentes')
plt.show()
```

___
## En que **COMUNA** ocurrieron mas siniestros

```python
plt.figure(figsize=(12,6))
sns.countplot(x='COMUNA', data=df_hechos)
plt.title('Cantidad de siniestros por Comuna')
plt.xticks(rotation=45)
plt.show()
```

___
## **PARTICIPANTES** con mayor presencia por cada comuna

```python
plt.figure(figsize=(12, 5))
sns.countplot(x='COMUNA', hue='VICTIMA', data=df_hechos)
plt.title('Cantidad de VICTIMAS por comuna')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='COMUNA', hue='ACUSADO', data=df_hechos)
plt.title('Cantidad de ACUSADOS por comuna')
plt.xticks(rotation=45)
plt.show()
```

___
## **HORA** En la que mas frecuntes suceden los siniestros

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='HH', data=df_hechos)
plt.xlabel('Hora')
plt.ylabel('Cantidad')
plt.title('Horas con Mayor Número de Siniestros')
plt.show()
```

___
## **RELACION** Entre HORA y PARTICIPANTES

```python
plt.figure(figsize=(18, 10))
sns.countplot(x='HH', hue='VICTIMA', data=df_hechos)
plt.title('Relación entre Hora y victimas')
plt.show()
```

Es interesante como se puede observar que en horarios de madrugrada no son tan presentes los peatones en los siniestros viales, y a la misma vez como las motos en horas de madrugrada pueden presentar mas riesgo por cuestiones como poca visibilidad desencadenando mas accidentes 

```python
plt.figure(figsize=(18, 10))
sns.countplot(x='HH', hue='ACUSADO', data=df_hechos)
plt.title('Relación entre Hora y acusados')
plt.show()
```
